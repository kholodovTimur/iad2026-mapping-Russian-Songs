```python
from pydantic import BaseModel, Field
from typing import Literal

from google import genai 

prompt = """
# ── Роль ────────────────────────────────────────────────────────
Ты — система NER для русскоязычных текстов песен.
Твоя задача: извлекать именованные географические объекты России.

# ── Вывод ───────────────────────────────────────────────────────
Отвечай строго в формате JSON, соответствующем схеме:
  {"places": [{"toponym": "...", "type": "..."}]}

# ── Типы ────────────────────────────────────────────────────────
Допустимые значения type:
  "улица"  — улица, проспект, переулок, шоссе, бульвар, набережная, площадь
  "метро"  — станция метро или метрополитен
  "район"  — район, округ, микрорайон, жилмассив
  "город"  — город, посёлок, село, деревня
  "страна" — государство, республика
  "другое" — река, гора, регион, лес, парк и всё остальное

# ── Обязательные правила ────────────────────────────────────────
1. toponym — дословно из текста. Падеж, регистр, написание не меняй.
2. Только именованные объекты. «улица», «река», «город» без имени — не включай.
3. Омонимия — проверяй контекст.
      «Волга» = автомобиль → не включай.
      «Волга» = река      → включай, type = "другое".
      «Урал»  = завод     → не включай.
      «Урал»  = регион    → включай, type = "другое".
4. Нет топонимов — верни {"places": []}.

# ── Примеры ─────────────────────────────────────────────────────
Текст:  «Еду по Тверской, выхожу на Пушкинской»
Ответ:  {"places": [
              {"toponym": "Тверской", "type": "улица"},
              {"toponym": "Пушкинской", "type": "метро"}
            ]}

Текст:  ««Волга» едет по МКАД»
Ответ:  {"places": [{"toponym": "МКАД", "type": "улица"}]}
              // «Волга» = автомобиль → пропускаем

Текст:  «Я в Питере снова»
Ответ:  {"places": [{"toponym": "Питере", "type": "город"}]}
              // разговорная форма — без нормализации
"""


class Place(BaseModel):
    toponym: str = Field(
        description="Топоним дословно из текста песни — без изменения "
                    "падежа, регистра и написания."
    )
    type: Literal[
        "улица", "метро", "район",
        "город", "страна", "другое"
    ] = Field(
        description="Категория географического объекта: "
                    "улица/проспект/бульвар/набережная/площадь — 'улица'; "
                    "станция метро — 'метро'; "
                    "район/округ — 'район'; "
                    "город/посёлок — 'город'; "
                    "государство — 'страна'; "
                    "всё остальное (река, гора, парк…) — 'другое'."
    )


class SongInfo(BaseModel):
    places: list[Place] = Field(
        description="Все именованные географические объекты России, "
                    "найденные в тексте песни. Пустой список — если не найдено."
    )

class AutoCoder:
    def __init__(self, api, prompt = prompt, scheme = SongInfo, model = 'gemini-3.1-flash-lite-preview'):
        self.api = api
        self.model = model
        self.prompt = prompt
        self.scheme = scheme
        self.client = genai.Client(api_key = api)

    def code(self, song_text):
        return self.client.models.generate_content(
            model = self.model,
            config = {
                "system_instruction": {self.prompt},
                "response_mime_type": "application/json",
                "response_json_schema": self.scheme.model_json_schema(),
            },
            contents = song_text
        )
```

```python
coder = AutoCoder()
```
